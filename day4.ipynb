```python
#visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
months=['Jan','Feb','Mar','Apr','May','Jun']
traffic=[150,2,180,220,250,210]
plt.plot(months, traffic)
plt.title('Monthly traffic website')
plt.show()
```

```python
month=['Jan','Feb','Mar','Apr','May','Jun']
product_a=[45,35,40,70,88,75]
product_b=[35,40,50,55,70,68]
plt.plot(month, product_a, marker='o', linestyle='-',color='blue', label='Product A')
plt.plot(month, product_b, marker='s', linestyle='--',color='red', label='Product B')
plt.xlabel('Month')
plt.ylabel('Monthly Revenue (in $1000)')
plt.title('Monthly Revenue Compariosn')
plt.legend()
plt.show()
```

```python
plt.plot(month, traffic, marker='o',linestyle='--', color='g')
plt.xlabel('Month')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Website Traffic')

plt.savefig('pythonlinechart')
plt.show()
```

```python
categories=['Category A','Category B','Category C', 'Category D']
values = [20,35,30,25]
plt.bar(categories, values, color=['red','g','black'], width=0.8)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot example')
plt.show()
```

```python
categories=['Category A','Category B','Category C', 'Category D']
values = [20,35,30,25]
plt.barh(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot example')
plt.show()
```
