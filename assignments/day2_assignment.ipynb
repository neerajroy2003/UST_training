```python
from functools import reduce

# assignment 1
# Local variables exist only inside the function.
# If we use them outside, Python gives NameError.

# assignment 2
# map
num1 = [1,2,3,4]
lst1 = list(map(lambda x: x+2, num1))
print(lst1)

# filter
num2 = [1,-2,3,4,0]
lst2 = list(filter(lambda x: x>0, num2))
print(lst2)

# reduce
num3 = [1,-2,3,4,5]
lst3 = reduce(lambda x, y: x*y, num3)
print(lst3)
```
