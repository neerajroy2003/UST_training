```python
#assignment
#with function and wihtout function do the matrix transpose
matrix=[[1,2,3,4],[6,7,3,4],[7,8,9,3]]
transpose=[]
for i in range(len(matrix[0])):
    row=[]
    for j in range(len(matrix)):
        row.append(matrix[j][i])
    transpose.append(row)
print(transpose)
```

```python
import numpy as np
matrix=[[1,2,3,4],[6,7,3,4],[7,8,9,3]]
tran=np.transpose(matrix)
print(tran)
```
