```python
#assessment
import math
def hcf(a,b):
    if a<b:
        a,b=b,a
    for i in range(1,min(a,b)):
        if a%i==0 and b%i==0:
            return i
        
def factorial(num):
    return math.factorial(num)
print(hcf(4,5))    
print(factorial(4))
```

```python
#map
squared2=[]
squared3=[]
lst=[1,2,3,4]
for num in lst:
    squared2.append(num**2)
    squared3.append(num**2)
print(squared2)
print(squared3)    
```

```python
#map
squared2=[]
squared3=[]
lst=[1,2,3,4]
squared2=list(map(lambda n:n**2,lst))
squared3=list(map(lambda n:n**3,lst))
print(squared2)
print(squared3) 
```

```python
#reduce
from functools import reduce
lst=[1,2,3,4]
def multiply(a,b):
    return a*b
print(reduce(multiply,lst))
print(reduce(lambda x,y:x*y,lst))
```

```python
#keyword arguments
def funct(**kwargs):
    print('name:{} age:{}'.format(kwargs['name'],kwargs['age']))
funct(name='Neeraj',age=15)
```

```python
#arbitary arguments
def funct(*args):
    for val in args:
        print('name:{}'.format(val))
funct('Neeraj','Ray')
```
