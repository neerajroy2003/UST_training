```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.random_sample(5)
plt.hist(data, color='green',edgecolor='yellow', bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram plot example')
plt.show()
```

```python
#Scatter plot
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,15,13,17,20]
plt.scatter(x,y,marker='v', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot example')
plt.show()
```

```python
#pie chart
categories=['housing','transportation','food','entertainment','utilities']
expenses=[1200,300,400,200,150]
plt.figure(figsize=(5,16))
plt.pie(expenses, labels=categories, autopct='%2.2f%%')
plt.title('Monthly expenses by Category')
plt.legend()
plt.show()
```

```python
#boxplot
data=[12,15,18,15,17,18,20,22,23,25,27,30,35,100,-34]
plt.boxplot(data)
plt.title('Boxplot example with sample data')
plt.xlabel('Data')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt
quarters=['01','02','03','04']
electronics=[1000,1200,1300,10500]
clothing=[5000,6000,7500,8200]
home_applicances=[3000,7500,8200,9000]
plt.figure(figsize=(10,6))
plt.stackplot(quarters, electronics, clothing, home_applicances, labels=['Electronics','Clothing','Home Applicances'])
plt.xlabel("Quarters")
plt.ylabel("Sales")
plt.title("Product Category")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2,y**2))
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)

fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].plot(x,y1,'r')
axs[0,0].set_title('sine wave')
axs[0,1].plot(x,y2,'b')
axs[0,1].set_title('Cosine wave')
axs[1,0].plot(x,y3,'g')
axs[1,0].set_title('Tangent wave')
axs[1,1].plot(x,y4,'m')
axs[1,1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```
