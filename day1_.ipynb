```python
#prime number
start=int(input("Enter start: "))
end=int(input("Enter end: "))
ar=[]
for i in range(start, end+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        ar.append(i)
print(ar)
        
    
```

```python
names=set()
while True:
    name=input("Enter name: ")
    if name=='exit':
        break
    elif name in names:
        print('Already exist')
    else:
        names.add(name)
print(names)
```
